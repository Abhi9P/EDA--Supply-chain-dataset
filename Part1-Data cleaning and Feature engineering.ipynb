### Data Cleaning and Feature Engineering 

```python
#combining customer first name and last name.

df['CustomerFullName'] = df['Customer Fname'] + ' ' + df['Customer Lname']
```

```python
dropped_col=['Days for shipping (real)','Order Item Total','Benefit per order','Days for shipment (scheduled)','Customer Fname','Customer Lname','Latitude','Longitude','Product Description','Customer Zipcode','Customer Street','Customer Email','Customer Password','Order Zipcode','Department Id','Order Customer Id','Order Item Discount Rate','Order Item Cardprod Id','Product Image','Product Category Id','Product Status','Product Price','Product Card Id']
df.drop(dropped_col,axis=1,inplace=True)
```

```python
df['order date (DateOrders)'] = df['order date (DateOrders)'].astype(str).str.split().str[0]
df['shipping date (DateOrders)'] = df['shipping date (DateOrders)'].astype(str).str.split().str[0]
```

```python
#Converting the data type into date time
df['order date (DateOrders)']= pd.to_datetime(df['order date (DateOrders)'])
df['order date (DateOrders)']
```

```python
#Converting the data type into date time
df['shipping date (DateOrders)']= pd.to_datetime(df['shipping date (DateOrders)'])
df['shipping date (DateOrders)']
```

```python
# Create a new column to calculate the delay in days for each shipment
df['Delay'] = (pd.to_datetime(df['shipping date (DateOrders)']) - pd.to_datetime(df['order date (DateOrders)'])).dt.days
```

**Sales**


```python
df['Sales'].unique()
```

```python
Sales_range = [0, 500, 1000, 1500, 2000]
Sales_labels = ['<500', '500-1000', '1000-1500', '1500-2000']

# Assign salary brackets to dataset
df['Sales Range'] = pd.cut(df['Sales'], bins=Sales_range, labels=Sales_labels)

# View the updated dataset
print(df.head())
```
